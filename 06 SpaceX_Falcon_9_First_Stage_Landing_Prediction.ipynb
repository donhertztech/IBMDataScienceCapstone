```python
# Coordinates for major nearby cities/landmarks
proximities = {
    'Cape Canaveral': (28.396837, -80.605659),
    'Orlando': (28.538336, -81.379234),
    'Miami': (25.761681, -80.191788),
    'Vandenberg AFB Coastline': (34.7420, -120.5724),
    'Los Angeles': (34.052235, -118.243683),
    'San Francisco': (37.773972, -122.431297)
}
```

```python
from math import radians, sin, cos, sqrt, atan2

def haversine(lat1, lon1, lat2, lon2):
    # Convert coordinates to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    radius = 6371  # Earth radius in km
    return c * radius
```

```python
# Get unique launch sites
launch_sites = spacex_df[['Launch Site', 'Lat', 'Long']].drop_duplicates()

# Calculate distances
distance_results = []
for _, site in launch_sites.iterrows():
    site_name = site['Launch Site']
    site_lat = site['Lat']
    site_lon = site['Long']

    for prox_name, (prox_lat, prox_lon) in proximities.items():
        distance = haversine(site_lat, site_lon, prox_lat, prox_lon)
        distance_results.append({
            'Launch Site': site_name,
            'Proximity': prox_name,
            'Distance (km)': round(distance, 2)
        })

# Create distance DataFrame
distance_df = pd.DataFrame(distance_results)
print(distance_df.pivot(index='Launch Site', columns='Proximity', values='Distance (km)'))
```

```python
# Create new map centered on launch sites
proximity_map = folium.Map(location=[30, -100], zoom_start=4)

# Add launch sites (from Task 1)
for _, site in launch_sites.iterrows():
    folium.Marker(
        location=[site['Lat'], site['Long']],
        popup=site['Launch Site'],
        icon=folium.Icon(color='blue', icon='rocket')
    ).add_to(proximity_map)

# Add proximity locations
for prox_name, (prox_lat, prox_lon) in proximities.items():
    folium.Marker(
        location=[prox_lat, prox_lon],
        popup=prox_name,
        icon=folium.Icon(color='green', icon='city' if 'Coastline' not in prox_name else 'water')
    ).add_to(proximity_map)

    # Add lines connecting launch sites to proximities
    for _, site in launch_sites.iterrows():
        folium.PolyLine(
            locations=[[site['Lat'], site['Long']], [prox_lat, prox_lon]],
            color='gray',
            weight=1,
            opacity=0.5
        ).add_to(proximity_map)

# Add distance labels
for _, row in distance_df.iterrows():
    site = launch_sites[launch_sites['Launch Site'] == row['Launch Site']].iloc[0]
    prox_lat, prox_lon = proximities[row['Proximity']]

    # Calculate midpoint for label placement
    mid_lat = (site['Lat'] + prox_lat) / 2
    mid_lon = (site['Long'] + prox_lon) / 2

    folium.Marker(
        location=[mid_lat, mid_lon],
        icon=DivIcon(
            icon_size=(150,50),
            icon_anchor=(75,25),
            html=f'<div style="font-size: 10pt">{row["Distance (km)"]} km</div>'
        )
    ).add_to(proximity_map)

proximity_map
```
